## In this project we analyze data on Thanksgiving dinner in the United States

#### First, we read the csv into a pandas dataframe and view the coulmn headers.

```python
import pandas as pd
data = pd.read_csv("thanksgiving.csv", encoding="Latin-1")
#print("shape =",data.shape, "\n", data.head(5))
print(data.columns)


```

#### Next, we filter out all the participants who do not celebrate thanksgiving. 

```python
#Filter on the column "Do you celebrate Thanksgiving?"
from pandas import Series
celeb_counts = pd.Series.value_counts(data["Do you celebrate Thanksgiving?"])
do_celeb = data["Do you celebrate Thanksgiving?"] == "Yes"
#print(do_celeb)
new_data = data[do_celeb]
#print(new_data)

#checking work (visual, and systematic)
#print(new_data["Do you celebrate Thanksgiving?"] == "Yes")
new_celeb_counts = pd.Series.value_counts(new_data["Do you celebrate Thanksgiving?"])
print(new_celeb_counts)
```

#### Here we'll examine who eats Tofurkey as their main dish, and the correlation that has with people who also eat gravy.

```python
#Establish dish and gravy vars
main_dishes = new_data["What is typically the main dish at your Thanksgiving dinner?"]
gravy_data = new_data["Do you typically have gravy?"]

#People that eat Tofurkey
main_dish_counts = pd.Series.value_counts(main_dishes)
#print(main_dish_counts)
tofur_true = main_dishes == "Tofurkey"
#print(tof_true)
tofur_main = main_dishes[tofur_true]
tofur_counts = pd.Series.value_counts(tofur_main)
#print(tof_counts)

#People that eat gravy
gravy_bool = gravy_data == "Yes"
gravy_true = gravy_data[gravy_bool]
#print(gravy_data[gravy_bool])
gravy_true_counts = pd.Series.value_counts(gravy_true)
#print(gravy_true_counts)

#People that eat Tofurkey and gravy
tofur_gravy = gravy_data[tofur_true] == "Yes"
print(tofur_gravy)
```

#### Next we'll examine how many people like at least one of three types of pie (apple, pumpkin, and pecan) with their Thanksgiving dinner.


```python
#Set up a series for each pie
apple_pies = new_data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple"]
#print(apple_pies)
pumpkin_pies = new_data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin"]
pecan_pies = new_data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan"]

#Convert each pie series to bools
apple_isnull = pd.isnull(apple_pies)
#print(apple_isnull)
pumpkin_isnull = pd.isnull(pumpkin_pies)
pecan_isnull = pd.isnull(pecan_pies)

#Determine who eats at least one of the three pies and the count overall
ate_pies = apple_isnull & pumpkin_isnull & pecan_isnull
print(pd.Series.value_counts(ate_pies))

```

```python
#Create var for age data, and establish counts for verification 
age = new_data["Age"]
age_null = pd.Series.isnull(age)
null_counts = pd.Series.value_counts(age_null)
print(null_counts)

#convert a str from the listed age options to an int
def age_split(some_str):
    if pd.isnull(some_str):
        return None
    str_split = some_str.split(" ")
    print(type(str_split))
    if len(some_str) <= 3:
        remove_plus = some_str.split("+")
        return int(remove_plus[0])
    return int(str_split[0])

int_age = pd.Series.apply(age, age_split)
print(pd.Series.describe(int_age))

###***HOW DO I GET RID OF ALL THOSE <class 'list>'S IN OUTPUT?***
```

These results, while an interesting exercise in applying functions across series, are not indicative of the actual ages of the participants; we took the lowest age from each grouping, and as such, it does not represent the actual participants. While still lacking, a more accurate method might be to use a random age from the range selected for each participant, which would create a more realistic ditribtion of ages.

```python
earning = new_data["How much total combined money did all members of your HOUSEHOLD earn last year?"]
earning_null = pd.isnull(earning)
null_earn_count = pd.Series.value_counts(earning_null)
print(null_earn_count)

def earn_split(earn_str):
    if pd.isnull(earn_str):
        return None
    split = earn_str.split(" ")
    earn = split[0]
    if earn == "Prefer":
        return None
    clean_earn = earn.replace(earn, ["$",","], '')
    clean_earn = int(clean_earn)
    return clean_earn

x = earning.apply(earn_split)
print(pd.Series.describe(x))
```

This data, while still lacking the veracity of specific income estimations associated with each participant, strikes me as more accurate than the age data, given that there are more options (10 compared to 4), and that the choice intervals seem to be smaller on a relative scale (between 1/20 and 1/8 of the total compared to between 1/6 and 7/30).
